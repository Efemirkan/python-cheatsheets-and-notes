# What is a Decorator?

A decorator is just a function that modifies or enhances another functionâ€™s behavior without permanently changing it.

##### Example without decorator syntax

```python
def greet():
    print("Hello!")

def add_politeness(func):
    def wrapper():
        print("Good morning!")
        func()
        print("Have a great day!")
    return wrapper

```

```python
greet()
```

```python
# manually wrap it
polite_greet = add_politeness(greet)
polite_greet()
```

Here, add_politeness is a decorator function.

##### Decorator Syntax (@)

```python
@add_politeness
def greet():
    print("Hello!")
```

```python
greet()
```

##### Decorators with Arguments

```python
def debug(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with {args} and {kwargs}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result}")
        return result
    return wrapper

@debug
def add(a, b):
    return a + b

add(3, 5)
```

##### Decorators that Take Arguments

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
```

Important because;

* Class methods, @classmethod, @staticmethod, Modify method behavior
* Async/Testing, @pytest.mark.asyncio, Mark async test functions
* AI / ML frameworks, @torch.no_grad(), Disable gradient tracking

##### Timing a Function

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def slow_addition():
    time.sleep(2)
    print("Done!")

slow_addition()
```

##### Decorators on Classes

```python
class MyClass:
    @staticmethod
    def static_example():
        print("No self or cls needed!")

    @classmethod
    def class_example(cls):
        print(f"Called from {cls.__name__}")
```

#### Logging Decorator

```python
import datetime

def log_calls(func):
    def wrapper(*args, **kwargs):
        time_now = datetime.datetime.now().strftime("%H:%M:%S")
        print(f"[{time_now}] Calling {func.__name__} with {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"[{time_now}] {func.__name__} returned {result}")
        return result
    return wrapper
```

```python
@log_calls
def multiply(a, b):
    return a * b

multiply(4, 5)
```

#### Error Handling Decorator

```python
def safe_run(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error in {func.__name__}: {e}")
            return None
    return wrapper
```

```python
@safe_run
def divide(a, b):
    return a / b

print(divide(10, 2))
print(divide(5, 0))
```

##### Combining Multiple Decorators

```python
@safe_run
@log_calls
def risky_divide(a, b):
    return a / b

risky_divide(10, 2)
risky_divide(10, 0)
```

#### Keep Metadata with (functools.wraps)

Without this, the wrapped function loses its original name and docstring.

```python
from functools import wraps

def log_calls(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper
```
