#### Core find/replace/search tools

```python
re.search(pat, text)     # first match anywhere → Match or None
re.match(pat, text)      # from start only (rarely what you want)
re.findall(pat, text)    # list of all non-overlapping matches (strings)
re.finditer(pat, text)   # iterator of Match objects (spans, groups)
re.sub(pat, repl, text)  # replace matches with repl (str or function)
re.split(pat, text)      # split by regex delimiter(s)
# For repeated use, compile once:
rx = re.compile(pat, flags=0)
```

```python
import re

email = input("What is your email?").strip()

if re.search("@", email):
    print("valid")

else:
    print("Invalid")
```

#### Bread-and-butter tokens

- .   any character except a new line
- \*   0 or more repetitions
- \+   1 or more repetitions
- ?   0 or 1 repetition
- {m} m repetitions
- {m,n} m-n repetitions

```python
email = input("What is your email?").strip()

if re.search(".+@.+", email):
    print("valid")

else:
    print("Invalid")
```

```python
email = input("What is your email?").strip()

if re.search(".+@.+.com", email):
    print("valid")

else:
    print("Invalid")
```

```python
email = input("What is your email?").strip()

if re.search(".+@.+.com", email):
    print("valid")

else:
    print("Invalid")
```

```python
import re

email = input("What's your email? ").strip()

if re.search(r".+@.+\.com", email):
    print("Valid")
else:
    print("Invalid")
```

we utilize the “escape character” or \ as a way of regarding the . as part of our string instead of our validation expression. 

- ^   matches the start of the string
- $   matches the end of the string or just before the newline at the end of the string

```python
email = input("What's your email? ").strip()

if re.search(r"^.+@.+\.com$", email):
    print("Valid")
else:
    print("Invalid")
```

- []    set of characters
- \[^]   complementing the set

```python
email = input("What's your email? ").strip()

if re.search(r"^[^@]+@[^@]+\.com$", email):
    print("Valid")
else:
    print("Invalid")
```

[^@]+ means any character except an @.

```python
email = input("What's your email? ").strip()

if re.search(r"^[a-zA-Z0-9_]+@[a-zA-Z0-9_]+\.com$", email):
    print("Valid")
else:
    print("Invalid")
```

[a-zA-Z0-9_] tells the validation that characters must be between a and z, between A and Z, between 0 and 9 and potentially include an _ symbol.

- \d    decimal digit
- \D    not a decimal digit
- \s    whitespace characters
- \S    not a whitespace character
- \w    word character, as well as numbers and the underscore
- \W    not a word character

```python
email = input("What's your email? ").strip()

if re.search(r"^\w+@\w.+\.(com|edu|gov|net|org)$", email):
    print("Valid")
else:
    print("Invalid")
```

- A|B     either A or B
- (...)   a group
- (?:...) non-capturing version
