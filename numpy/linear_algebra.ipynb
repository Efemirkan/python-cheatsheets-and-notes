```python
import numpy as np
```

**Vectors and Matrices**

```python
# 1D Vector
x = np.array([1, 2, 3])
print(x)
```

Used in ML; A feature vector, A weight vector, A gradient

```python
# 2D Matrix
X = np.array([[1, 2, 3],
              [4, 5, 6]])
print(X)
```

In ML; Rows as samples, Columns as features

```python
# Shape
print(X.shape)
```

**Transpose — np.transpose() or .T**

```python
# Transpose swaps rows and columns.
X_T = X.T
print(X_T)
```

Used in ML for:

- Normal equations
- Gradient computation
- Covariance matrices

**Dot Product & Matrix Multiplication**

```python
# Vector dot product
x = np.array([1, 2, 3])
w = np.array([0.1, 0.2, 0.3])

y = np.dot(x, w)
print(y)
```

Mathematically:

$ y= x^Tw$

```python
# Matrix × Vector (Linear model)
X = np.array([[1, 2],
              [2, 3],
              [3, 4]])

w = np.array([0.5, 1.0])

y_hat = X @ w
print(y_hat)
```

$\hat{y} = Xw$

```python
# Matrix × Matrix (Neural network layers)
W1 = np.array([[0.2, 0.3],
               [0.4, 0.5]])

W2 = np.array([[1.0],
               [2.0]])

Z = W1 @ W2
print(Z)
```

**Identity Matrix — np.eye()**

```python
# Identity matrix = matrix version of number 1

I = np.eye(3)
print(I)
```

Property

A @ I == A

Used in ML; Regularisation, Numerical stability

**Matrix Inverse — np.linalg.inv()**

```python
A = np.array([[4, 7],
              [2, 6]])

A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
# Check
print(A @ A_inv)
```

Output ≈ Identity matrix.

Used in closed-form linear regression:

$w = (X^T X)^{-1} X^T y$

**Determinant — np.linalg.det()**

```python
A = np.array([[1, 2],
              [3, 4]])

print(np.linalg.det(A))
```

Used in ML:

- Check if matrix is invertible
- Detect degenerate systems

If determinant = 0 means **no inverse**

**Norms — np.linalg.norm()**

Norm = vector length (magnitude).

```python
# L2 norm (Euclidean)
x = np.array([3, 4])
print(np.linalg.norm(x))
```

Used in ML:

- Regularisation (L2)
- Distance metrics
- Gradient magnitude

**Eigenvalues & Eigenvectors — np.linalg.eig()**

```python
A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

Used in ML:

- PCA
- Spectral clustering
- Stability analysis

**Singular Value Decomposition (SVD) — np.linalg.svd()**

```python
X = np.array([[1, 2],
              [3, 4],
              [5, 6]])

U, S, Vt = np.linalg.svd(X)

print("U:\n", U)
print("S:\n", S)
print("Vt:\n", Vt)
```

Used in ML:

- PCA
- Dimensionality reduction
- Latent semantic analysis (NLP)

**Diagonal Matrix — np.diag()**

```python
v = np.array([1, 2, 3])
D = np.diag(v)
print(D)
```

Used in ML:

- Covariance matrices
- Weight decay
- Spectral methods

**Example**

```python
# Data
X = np.array([[1, 1],
              [1, 2],
              [1, 3]])  # bias term included

y = np.array([1, 2, 3])

# Normal Equation
w = np.linalg.inv(X.T @ X) @ X.T @ y

print("Weights:", w)
```
