```python
import numpy as np
```

**Vectors and Matrices**

```python
# 1D Vector
x = np.array([1, 2, 3])
print(x)
```

Used in ML; A feature vector, A weight vector, A gradient

```python
# 2D Matrix
X = np.array([[1, 2, 3],
              [4, 5, 6]])
print(X)
```

In ML; Rows as samples, Columns as features

```python
# Shape
print(X.shape)
```

**Transpose — np.transpose() or .T**

```python
# Transpose swaps rows and columns.
X_T = X.T
print(X_T)
```

Used in ML for:

- Normal equations
- Gradient computation
- Covariance matrices

**Dot Product & Matrix Multiplication**

```python
# Vector dot product
x = np.array([1, 2, 3])
w = np.array([0.1, 0.2, 0.3])

y = np.dot(x, w)
print(y)
```

Mathematically:

$ y= x^Tw$

```python
# Matrix × Vector (Linear model)
X = np.array([[1, 2],
              [2, 3],
              [3, 4]])

w = np.array([0.5, 1.0])

y_hat = X @ w
print(y_hat)
```

$\hat{y} = Xw$

```python
# Matrix × Matrix (Neural network layers)
W1 = np.array([[0.2, 0.3],
               [0.4, 0.5]])

W2 = np.array([[1.0],
               [2.0]])

Z = W1 @ W2
print(Z)
```

**Identity Matrix — np.eye()**

```python
# Identity matrix = matrix version of number 1

I = np.eye(3)
print(I)
```

Property

A @ I == A

Used in ML; Regularisation, Numerical stability
