# **Array Creation & Shape**

```python
import numpy as np
```

**np.array() - Create an array from a Python list**

```python
# 1D array (vector)
x = np.array([1, 2, 3, 4])
print(x)
```

- A feature vector

- A weight vector

- A single data sample

```python
# 2D array (matrix)
X = np.array([[1, 2, 3],
              [4, 5, 6]])
print(X)
```

- Rows = samples

- Columns = features

**np.zeros() - Create an array full of zeros**

```python
# Vector of zeros
w = np.zeros(5)
print(w)
```

- Initial weight vector

- Gradient storage

```python
# Matrix of zeros
W = np.zeros((3, 4))
print(W)
```

**np.ones() - Create an array full of ones**

```python
b = np.ones(3)
print(b)
```

- Bias terms

- Normalisation baselines

```python
# 2D version
B = np.ones((2, 3))
print(B)
```

**np.empty() - Allocate memory without filling values**

```python
A = np.empty((2, 3))
print(A)
```

Values are random garbage in memory. Used for fast performance.

- High-performance simulations

- When you overwrite values immediately

**np.arange() - Like Python range, but for arrays**

```python
t = np.arange(0, 10)
print(t)
```

- Index arrays

- Time steps

- Numeric sequences

```python
x = np.arange(0, 10, 2) # 2 step size
print(x)
```

**np.linspace() - Evenly spaced numbers**

```python
x = np.linspace(0, 1, 5)
print(x)
```

When you want exact number of points. important for plotting & ML

- Activation function plots

- Loss curves

- Continuous simulations
