```python
import numpy as np
```

**np.min() and np.max() - Find range**

```python
x = np.array([10, 20, 50, 80])

print(np.min(x))
print(np.max(x))
```

```python
X = np.array([[1, 200],
              [2, 300],
              [3, 400]])

print(np.min(X, axis=0))
print(np.max(X, axis=0))
```

**np.ptp() - Peak-to-Peak (range)**

```python
x = np.array([10, 20, 50, 80])
print(np.ptp(x))
```

```python
max(x) - min(x)
```

**Minâ€“Max Normalisation**

```python
x = np.array([10, 20, 50, 80])

x_norm = (x - np.min(x)) / np.ptp(x)
print(x_norm)
```

**np.percentile() - Robust statistics**

```python
x = np.array([10, 12, 15, 18, 100])

print(np.percentile(x, 50))  # median
print(np.percentile(x, 90))
```

**np.round(), np.floor(), np.ceil()**

```python
x = np.array([1.2, 2.5, 3.7])
print(np.round(x))
```

```python
print(np.floor(x))
```

```python
print(np.ceil(x))
```

**np.vstack() - np.hstack()**

```python
# Vertical stack - add samples
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6]])

X = np.vstack((A, B))
print(X)
```

```python
# Horizontal stack - add features
y = np.array([[10],
              [20],
              [30]])

X_new = np.hstack((X, y))
print(X_new)
```

**np.concatenate()**

```python
X = np.concatenate((A, B), axis=0)
```

**Standardisation**

```python
# (Mean = 0, Std = 1)
x = np.array([10, 20, 30, 40])

x_std = (x - np.mean(x)) / np.std(x)
print(x_std)
```
