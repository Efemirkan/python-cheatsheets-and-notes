```python
os.makedirs("results", exist_ok=True)
```

**File & Folder Existence Checks**

```python
os.path.exists("data.csv")
os.path.isfile("data.csv")
os.path.isdir("data")
```

```python
if not os.path.exists("models"):
    os.makedirs("models")
```
